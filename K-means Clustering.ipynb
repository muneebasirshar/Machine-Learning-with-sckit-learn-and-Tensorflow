clustering is typically an unsupervised learning. Labels are generally not known at the begining.
The following steps are followed:
1. select random points as centriods. 
2. find the distance of each point from each of these centriods
3. assign each point to the cluster centriod whose distance is minimum to its position.
4. after all the points have been assigned to a centriod find mean for each cluster
5. the cluster mean is the new centriod.
Repeat the steps 2-5 untill convergence

The quality of a cluster depands on the variance captured by the cluster

```python
def elbow_method():
    explained_variances = []
    Ks = list(range(3, 30))
    for K in Ks:
        kmeans_model = KMeans(K, random_state=1)
        y_pred = kmeans_model.fit_predict(data_frame)
        explained_variance  = kmeans_model.score(data_frame) 
        explained_variances.append(explained_variance)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(Ks, np.exp(explained_variances), marker='o')
    plt.xlabel('number of clusters')
    plt.ylabel('variance captures')
    plt.title('optimal number of cluster members')
    plt.show()
    
elbow_method()
```
